```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('charts.csv')
df.head(10)
```

```python
df_global_top200 = df[(df['region']=='Global') & (df['chart']=='top200')]
```

```python
df_global_top200 = df_global_top200.sort_values(by='date')
```

```python
df_global_top200
```

```python
df_global_top200.info()
```

```python
df_bts_global_top200 = df_global_top200[(df_global_top200['artist']=='BTS')]
```

```python
df_bts_global_top200
```

```python
df_bts_global_top200_title_count = df_bts_global_top200['title'].nunique()
df_bts_global_top200_title = df_bts_global_top200['title'].unique()
```

```python
print(df_bts_global_top200_title_count)
print(df_bts_global_top200_title)
```

BTS는 2017년 1월 ~ 2021년 12월 까지 총 76개의 곡들을 global top200에 차트인 시킨걸 볼 수 있다. </br>
이 곡들은 솔로, 다른 가수와의 듀엣을 제외한 bts가 부른 곡들 or bts곡들에 다른 가수들이 피쳐링한 곡들만 나와있는 숫자이다

```python
df_bts_top3=df_bts_global_top200[(df_bts_global_top200['title'].isin(['Boy With Luv (feat. Halsey)', 'Butter', 'Dynamite']))]
```

```python
df_bts_top3
```

```python
df_bts_global_top200['title'].value_counts()
```

top200에 들어간 곡들 중에 200일 이상 차트인 했던 곡들은</br>
2017년 1월~ 2021년12월20일까지의 기간들은</br>

Dynamite 487일</br>
Boy With Luv (feat. Halsey) 306일</br>
Butter 214일</br>

이 세곡을 기준으로 차트 그래프를 그려보게되면

```python
plt.figure(figsize=(20, 20))
sns.scatterplot(data=df_bts_top3, x='date', y='streams', hue='title')
plt.yscale('log')
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.lineplot(data=df_bts_top3, x='date', y='rank', hue='title')

# y축의 방향을 반대로 설정
plt.gca().invert_yaxis()

plt.title('Boy With Luv (feat. Halsey) & Dynamite & Butter Charts')
plt.show()
```

그래프는 BTS의 노래 중 <span style="color:blue">작은 것들을 위한 시</span>, <span style="color:orange">Dynamite</span>, <span style="color:green">Butter</span>에 Global Charts top200의 그래프이다.</br>
세 곡 모두 시간이 지날수록 내려가는걸 볼 수 있지만 특이한 점은 신곡이 발매를 하게되면 그전에 나왔던 곡들의 순위가 급상하는걸 볼 수 있다.</br>
특히 <span style="color:blue">작은 것들을 위한 시</span>의 차트가 흥미로운데 <span style="color:orange">Dynamite</span>발매 할 때 쯤 순위는 급격하게 상승하는걸 볼 수 있고 <span style="color:green">Butter</span>의 발매 전까지 차트에 없다가 <span style="color:green">Butter</span>발매 이후 다시 차트인 하는걸 볼 수 있다</br>
(여기서 선이 그어진 이유는 lineplot는 연속적으로 그리기 때문 위에 스트리밍 그래프를 보면 이때 스트리밍 scatter이 찍히지 않았다는걸 볼 수 있다.)



<span style="color:orange">Dynamite</span>와 <span style="color:green">Butter</span>의 그래프를 보게되면 <span style="color:green">Butter</span>가 발매한 이후로 꾸준히 그래프의 등락률이 비슷하게 보이는게 매우 흥미롭다. 비록 순위 차이는 나고 있지만 21년 12월쯤을 제외하고는 계속 같은 그래프의 모습을 띄고있다.</br>

이 그래프를 보며 본인은 BTS의 팬층이 탄탄하다고 생각이 들었다 많은 팬들이 두 곡 모두 듣고 대중들이 신곡을 듣게되면 자연스럽게 대중들이 듣게되는 신곡인 <span style="color:green">Butter</span>의 순위는 올라가고 이와 탄탄한 팬층으로 인해 <span style="color:orange">Dynamite</span>의 그래프도 비슷한 그래프를 보인다고 생각한다.</br>

따라서 그 이후에 나온 BTS 멤버들의 솔로곡이 나온 이후의 BTS 곡들의 차트도 비슷한지 궁금하기 때문에 2022년의 차트 데이터를 받아서 비교해봐야겠다.</br>

최대한 k-pop과는 멀어보이는 나라 위주로 뽑아 봤고 이 나라들에 대한 BTS 차트를 분석해 보겠습니다.

```python
df_bts_top200 = df[(df['chart'] == 'top200') & (df['artist']=='BTS')]
```

```python
df_bts_Saudi_Arabia_top200 = df_bts_top200[df_bts_top200['region']=='Saudi Arabia']
df_bts_Lithuania_top200 = df_bts_top200[df_bts_top200['region']=='Lithuania']
df_bts_Morocco_top200 = df_bts_top200[df_bts_top200['region']=='Morocco']
df_bts_Canada_top200 = df_bts_top200[df_bts_top200['region']=='Canada']
df_bts_Nicaragua_top200 = df_bts_top200[df_bts_top200['region']=='Nicaragua']
df_bts_New_Zealand_top200 = df_bts_top200[df_bts_top200['region']=='New Zealand']

df_bts_Continents_top200 = df_bts_top200[df_bts_top200['region'].isin(['Global', 'Saudi Arabia', 'Lituania', 
                                                                        'Morocco', 'Canada', 'Nicaragua', 'New Zealand'])]
```
