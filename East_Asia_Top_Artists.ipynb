```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('east_asia_top_artists.csv')
df.head()
```

```python
df.info()
```

```python
df = df.drop(['Unnamed: 0'], axis=1)
df
```

```python
df = df.drop_duplicates(['artist_name'])
df
```

<p style='font-weight:bold'> 중복 아티스트 제거</p>

```python
df_k = df[df['genres'].str.contains('k-|korea')]
df_j = df[df['genres'].str.contains('j-|japan')]
df_other = df[~df['genres'].str.contains('k-|korea|j-|japan')]
```

```python
df_count=[len(df_k), len(df_j), len(df_other)]
labels=['K-pop', 'J-pop', 'Other']
plt.pie(df_count, labels=labels, autopct='%1.1f%%', startangle=0)
plt.show()
```

## 1. 종류에 따른 비율

 - J-pop이 50% k-pop이 26%의 비율을 보이며 K-pop의 위상이 높아졌음에도 아직 Spotify에서 J-pop가수들의 비율이 더 높다.
 - 여기서 Other이라고 지칭한 이유는 중국가수들이 아닌 자메이카나 서양 가수들이 많이 나왔기 때문에 Other이라고 지칭함
 - 그렇기 때문에 주로 K-pop과 J-pop을 사용해서 데이터 분석을 해보겠음

```python
df_other
```

```python
df_K_J = pd.concat([df_k, df_j], axis=0).reset_index()
df_K_J.loc[df_K_J['genres'].str.contains('k-|korea'), ['genres']] = 'K-pop'
df_K_J.loc[df_K_J['genres'].str.contains('j-|japan'), ['genres']] = 'J-pop'
```

```python
df_K_J
```

```python
df_K_J = df_K_J.drop('index', axis=1)
df_K_J = df_K_J.sort_values(by='followers' ,ascending=False)
```

```python
plt.figure(figsize=(20, 10))
sns.barplot(data = df_K_J.head(50) , x='artist_name', y='followers', hue='genres')
plt.xticks(rotation=75, fontsize=11)
plt.show()
```

## 2. 팔로워수 상위 50팀
 * 이전에 본 그래프에서 J-pop 가수들이 더 많은 비율을 차지했지만 상위 50팀의 가수를 보게되면 압도적으로 K-pop 가수들이 많은걸 보여준다
 * 특히 BTS는 2위 BLACKPINK와도 차이가 많이 보이며 BTS 개개인 멤버들도 상위권에 분포되어있는걸 보여줌

```python
df_K_followers = df_k['followers'].sum()
df_J_followers = df_j['followers'].sum()
df_bts_blackpink = df_K_followers - df_k['followers'][:2].sum()
```

```python
print('K-pop followers :', df_K_followers)
print('J-pop followers :', df_J_followers)

print('BTS,BLCKPINK 제외한 K-pop followers :',df_bts_blackpink)
```

```python
followers_count = [df_K_followers, df_J_followers]
followers_count_2 = [df_bts_blackpink, df_J_followers]
labels = ['K-pop', 'J-pop']


plt.subplot(1,2,1)
plt.pie(followers_count, labels=labels, autopct='%1.1f%%', startangle=0)
plt.title('Followers')


plt.subplot(1,2,2)
plt.pie(followers_count_2, labels=labels, autopct='%1.1f%%', startangle=0)
plt.title('Excluding BTS and BLACKPINK')

plt.show()
```

**팔로워의 총 합을 나타내주는 그래프이다.**
* BTS와 BLACKPINK가 압도적으로 높기 때문에 당연히 K-pop의 팔로워 수가 더 높게 나왔다.
* 두번째 그래프는 BTS와 BLACKPINK의 팔로워 수를 제외한 그래프인데 그래도 K-pop의 팔로워가 더 높다
* 따라서 스포티파이에선 K-pop의 위상이 더 높지만, 아직 J-pop의 가수들이 더 많이 있는걸로보아 지금까지는 J-pop이 더 유명했던걸로 보임
